## 2. 관심데이터를 그룹화 하고 요약하기
### Heart csv 파일경로 지정

데이터 그룹화 및 요약
- 데이터를 특정 열을 기준으로 데이터를 그룹별로 나누고, 해당 그룹에 대한 통계를 내기 위해서 df_hf.groupby('col'.agg(func))를 기본구조로 groupby와 aggregating함수를 사용할 수 있다.
- groupby는 데이터를 특정 열의 값 별로 나누며, aggregating 함수는 그룹화된 데이터에 대한 통계적 요약을 계산할 수 있다.
- Aggregating 주요 함수
    - mean()
    - sum()
    - min()/max()
    - median()
    - std()
    - count()
