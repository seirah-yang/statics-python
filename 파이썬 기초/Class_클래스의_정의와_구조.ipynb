## 1. 클래스 정의
클래스는 객체지향적 언어들이 가지고 있는 특성으로 인스턴스(객체)를 만들기 위한 도구이다.

- 병원에서의 상황을 생각해보면 진료를 받기 위해 병원에 방문한 사람들은 모두 환자라고 불리지만, [환자들은 모두 다른 **속성(나이, 성별, 몸무게)과 다른 질병적 특징**을 가지고 있다. ]

- EMR에 작성시 우리는 만들어진 틀에 환자들의 속성을 기입해 넣는다.

    - 클래스: 환자들이 가지고 있을 특성을 포괄적으로 정의해 놓은 표(form)
    - 인스턴스: 환자들의 정보를 기입해 놓은 차트들


클래스의 장점 :
- 코드의 재사용률을 높인다.
- 객체별로 다르게 정의될 수 있는 부분은 구분해서 저장할 수 있다.


## 클래스 구조 이해하기
- class라는 명령어를 통해 클래스의 정의를 정의
- 클래스를 부를때 사용하게 될 클래스명을 입력
- 보통은 대문자로 시작하는 이름을 클래스에 부여
- 클래스의 해당하는 함수들은 일반적인 함수를 정의하는 방법과 유사하지만, self라는 입력을 꼭 함수의 파라미터로 작성해야 함
- self는 클래스로 생성된 인스턴스 자체를 의미
- 사전에 정의된 클래스 내부 변수와 함수를 호출하는데 사용된다.

        class 클래스 명:
                def 함수명 (self, ...):
                        [함수코드]

```python
# 클래스 정의 예시
class  Hospital:
    def __init__(self, name):
        self.name = name

    def print_name (self):
        print(self.name, "입니다.")
        return

    def count_patients (self, patients):
        return len(patients)

hospital_a = Hospital(name = "세브란스병원")
hospital_a.print_name()
```

```python
hospital_a.count_patients(patients=["A","V","B","D","E","N","C"])
```

- 초기화(__init__) 함수 ➞  함수의 입력으로 환자번호(pid),
    환자 성명(name), 나이(age), 성별(sex), 과(dept), 질병(disease)를 받고 , self를 통해 클래스 내의 속성으로 각각의 항목을 저장
- 입원상태는 False로, 외래 상태는 True로 초기화하고, 입원 날짜와 퇴원 날짜는 모두 None값(아무 값도 없는 상태)으로 설정
- 클래스의 입원설정(set_admission)과 외래설정(set_outpatient) 메소드 = 날짜를 입력으로 받는다.
    - set_admission에서는 입원상태를 True로 외래상태를 False로 변경, 동시에 입력받은 날짜를 self.admission_date에 부여
- set_outpatient는 이와 반대로 설정
    -  status라는 이름을 가진 메소드는 환자 클래스에 적용되어 있는 환자 속성을 출력해주는 코드로 환자의 모든 상태를 string형태로 변환하여 출력
- self.sex 가 남이라는 문자라면 self.sex뒤에 ‘성’이라는 글자를 추가하여 남성이 되고, self.sex 가 남이 아니라면 여성을 출력하는 것으로 설정
- 입원 상태에 대한 출력도 입원상태(self.admission)가 True이라면 ‘입원’, 아니면 ‘외래’로 출력되도록 정의

```python
class patient:
    def __init__(self,pid,name,sex,age,dept,disease):
        #환자 기본정보
        self.pid = pid
        self.name=name
        self.sex=sex
        self.age=age
        self.dept=dept
        self.disease=disease

        #입원정보
        self.admission = False
        self.admission_date = None

        #외래정보
        self.outpatient = True
        self.discharge_date = None

    def set_admission(self, date):
        self.admission = True
        self.outpatient = False

        self.admission_date = date

        print(f"{self.name}님은 {self.admission_date}부로 입원환자로 전환되셨습니다.\n")

    def set_admission(self,date):
        self.admission = False
        self.outpatient = True

        self.discharge_date = date

        print(f"{self.name}님은 {self.discharge_date}부로 외래환자로 전환되셨습니다.\n")

    def status(self):
            print(f"{self.name}님은 {self.age}세 {self.sex + '성' if self.sex == '남' else '여성'}으로 현재 {'입원' if self.admission == True else '외래'} 상태입니다.")
            print(f"{self.dept}에서 {self.disease}을 진단 받으셨습니다.\n")
```

```python
patientList = []

#enumerate를 활용한 반복문

for i, person in enumerate([{"name":"홍길동", "age":78,
                                                                "sex":"남", "dept":"내분비내과", "disease":"당뇨병"},
                                                            {"name":"홍길순", "age":75,
                                                                "sex":"여", "dept":"심장내과", "disease":"협심증"}]):
    #클래스 인스턴스 정의  후 리스트에 추가
    patientList.append(patient(pid = i+1,
                                                name = person["name"],
                                                age = person["age"],
                                                sex = person["sex"],
                                                dept = person["dept"],
                                                disease = person["disease"]))

    #첫번째 환자 상태 출력
    patientList[0].status()
```

patient 클래스를 통해 각각의 instance를 만들어 두고 patientList에 추가해준다.

첫번째 환자(홍길동)의 상태를 출력하는 코드와 두번째 환자의 상태를 외래에서 입원으로 전환하고 상태를 출력한다.

```python
patientList[1].set_admission("2025년 08월 19일")
```

```python
patientList[1].status()
```
