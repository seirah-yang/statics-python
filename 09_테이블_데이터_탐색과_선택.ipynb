## 테이블 데이터 선택
데이터 분석과 탐색이 끝났다면 데이터 선택을 통해 데이터프레임에서 특정 데이터를 선택하여 특정 부분에 집중하거나 불필요한 정보를 제외시킬수 있다.

```python
df = pd.DataFrame([[1,2], [4,5], [7,8]],
                  index=['cobra','viper','sidewinder'],
                  columns=['max_speed', 'shield'] )

df
```

시리즈 & 데이터프레임 반환

df['column_name']
단일열을 선택하여 시리즈로 반환할 수 있다.

예) max_spped열을 선택하면 df['max_speed']를 시리즈로 반환할 수 있다.

```python
df['max_speed']
```

```python
#데이터프레임 반환
df[['max_speed','shield']]
```

```python
# 행 선택 후 데이터 선택적 불러오기
#loc 메서드: label-based indexing
# df.loc[row_label]은 row_label을 기준으로 행을 선택

df.loc['viper']
```

```python
df.loc[['viper', 'sidewinder']]
```

```python
df.loc['cobra', 'shield']
```

```python
df.loc['viper', 'max_speed']
```

```python
df.loc['cobra':'viper', 'max_speed']
```

```python
#iloc 메서드: integer-based indexing
#인덱스 기준으로 행을 선택하는데, 이는 행의 순서를 나타내는 것으로 행을 선택하려면 df.iloc[0]를 사용할 수 있다.
type(df.iloc[0])
```

```python
df.iloc[[0]]
```

```python
df.iloc[[0,1]]
```

```python
df.iloc[[0,1]]
```
