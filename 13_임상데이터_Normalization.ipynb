## 데이터 정규화 (Normalization)
### 정규화(Normalization)
: 특성(Feature)간의 크기 차이가 다를 때 사용하는 방법
- 데이터셋의 다양한 값들 중에서 너무 큰 값이나 작은 값 때문에 모델의 학습 성능이 저하되는 것을 방지
- 데이터 스케일링 (Scaling)을 통해 모든 데이터를 일정한 범위로 "표준화"하여 변환

```python

import kagglehub

# Download latest version
path = kagglehub.dataset_download("whenamancodes/heart-failure-clinical-records")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import os


path = "/kaggle/input/heart-failure-clinical-records"
print(os.listdir(path))
```

```python
df_hf = pd.read_csv(os.path.join(path, 'Heart Failure Clinical Records.csv'))
```

```python
df_hf.head(5)
```

```python
from sklearn.preprocessing import MaxAbsScaler

# df_hf 데이터프레임 'age'와 'platelets' 상위 5개 데이터 출력
dt1= df_hf[[ 'age', 'platelets' ]].head(5)
dt1
```

```python
abs = MaxAbsScaler()
# fit 함수로 최대 절대값 계산
abs.fit(dt1)

# transform 함수로 dt1 데이터 스케일링
scaled_dt= abs.transform(dt1)

# 데이터프레임으로 출력
dt_scaled = pd.DataFrame(scaled_dt, columns=dt1.columns)
dt_scaled
```

```python
from sklearn.preprocessing import MinMaxScaler
# MinMaxScaler 객체 생성
minmax = MinMaxScaler()

# fit_transform 함수를 사용하여 dt1 데이터 스케일링
dt_norm = pd.DataFrame(minmax.fit_transform(dt1), columns=dt1.columns)

# 스케일링된 데이터프레임 출력
dt_norm
```

```python
#원본 데이터프레임 확인
dt1= df_hf[[ 'age' , 'platelets' ]].head(5)
dt1
```

```python
# StandardScaler 불러오기
from sklearn.preprocessing import StandardScaler

# StandardScaler 객체 생성 후 std_s에 할당
std_s = StandardScaler()
std_s

# 평균: 0 , 표준 편차: 1
df_std = pd.DataFrame(std_s.fit_transform(dt1), columns=dt1.columns)
df_std
```

```python
from sklearn.preprocessing import RobustScaler

# RobustScaler로 객체 생성
scaler = RobustScaler()

# fit_transform 함수로 dt1의 데이터 중앙값과 IQR 계산 후 데이터 스케일링
df_robust = pd.DataFrame(scaler.fit_transform(dt1), columns=dt1.columns)
# 데이터프레임 출력
print (df_robust)
```
