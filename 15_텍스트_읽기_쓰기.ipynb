### 1) 파일 읽고 쓰기
의료데이터를 다루는 과정에서 직접 raw데이터를 읽어오고, 저장이 필요할 때, 메모리에 저장된 데이터를 영구적으로 저장하거나 파일에 데이터를 저장해 공유하는 목적으로도 파일 읽고, 쓰기가 필요하다.

(1) 파일열기 : open() 함수 및 기본모드
- 기본구조: file=open("example_file.txt", "mode", encoding="utf-8")
    - 파일경로를 포함한 파일이름
    - 읽기, 쓰기, 바이너리 등 목적에 맞는 모드 지정하는 mode
    - 인코딩을 설정할 수 있는 encoding

- encoding은 기본적으로 파이썬을 설치한 운영체제의 시스템 인코딩을 따라가기 때문에 특별한 경우가 아닌 이상 생략이 가능

- 'w' write 쓰기모드 , 'r' 읽기모드, 'a' 추가모드, 'b' 바이너리모드('rb', 'ab', 'wb')


(2) 파일 닫기: close , with
- 파일작업이 완료 된 후 열린 파일을 닫을 때에는 file.close()
- 파일 닫는 작업이 실행되지 않을 경우 데이터 손실 및 문제가 발생할 수 있다.
- with 구문을 사용하면 코드블록의 실행이 끝날때 자동으로 파일을 닫아 주기 때문에 `fille.close()`가 따로 필요가 없다.

```python
# open 함수 사용
# 새로운 빈 파일을 write모드로 생성
file = open("example_file.txt", "w", encoding="utf-8")
#file.close()
```

```python
file.write("의료인을 위한 실전 의료영상 데이터 다루기")
file.close()
```

```python
# example_file.txt을 read모드로 읽어보기
# with 구문사용시
with open("example_file.txt", "r") as file:
    file.write("의료인을 위한 실전 의료영상 데이터 다루기")
    file.close()
```

# 3. 기존 파일문자 변경 및 파일 관리
## 기존파일 수정
파일을 저장하는 과정에서 text자체의 특성인 replace함수를 함께 사용해 특정 문자 혹은 문자열 앞에 원하는 내용을 추가 ➜ 다른 문자로 교체

### ▶ replace 함수:
- string 타입의 내장함수
- 파일의 특정 기호나 문자 변경
기본 구조로: variable.replace("old_substring", "new_substring") - old_substring = 바꾸고 싶은 문자열
- new_substring = 새로운 문자열

```python
# 단어번경
# 기존파일 내용 content에 저장
with open("example_file.txt", "r", encoding="utf-8") as file:
    content = file.read()
```

```python
# 'Version' 을 '버전' 으로 변경
content = content.replace("Version", ",버전")
```

```python
# 변경된 내용을 파일에 다시 저장
with open("example_file.txt", "w", encoding="utf-8") as file:
    file.write(content)

# 변경 후 문자열:
print(content)
```

```python
# 기존파일 내용 content에 저장
with open("example_file.txt", "r", encoding = 'utf-8') as file:
    content = file.read()

# 기존파일 문자열:
print("변경전:",content)
```

```python
# 기존파일 문자열에 특정 부분 다른 문자열로 변경
update = content.replace("의료인을 위한 실전 의료영상 데이터 다루기",
                         "의료인을 위한 하루에 10분씩 100일 실전 의료영상 데이터 다루기")

# 변경된 내용을 파일에 다시 저장
with open("example_file.txt", "w", encoding='utf-8') as file:
    file.write(update)

# 변경 후 문자열:
print("변경후:", update)
```

## o for 문과 strip() 함수로 텍스트 읽어오기 및 문자열 가공
- 파일의 내용을 텍스트로 읽어올땐 for loop 을 사용한다.
- 텍스트에 원치않는 공백이나 문자가 있다면 이를 strip()함수를 통해 가공할수 있다.

- open() 사용하여 파일 객체를 file 변수에 저장한 후, for loop을 사용하여 파일의 각 라인을 한줄씩 읽어옴
- strip() 함수를 통해 각 줄의 앞뒤 공백 (또는 원하는 문자)를 제거하고 내용을 출력

```python
with  open("example_file.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
text = "___의료인책, 100일!___"
# 공백이 아닌 '_' 문자열을 정리할시:
cleaned_text = text.strip("_")
print(cleaned_text)
```

## o 파일 이름변경 및 삭제

- 파이썬의 라이브러리인 os 모듈을 사용하면 해당 작업들을 진행할 수 있다.
