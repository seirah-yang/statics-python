```python
#df1와 df2 'diabetes'열기준 데이터프레임 합치기
# 'age'와 'creatinine_phosphokinase'의 mean (평균)
df1 = df.groupby('diabetes')[["age", "creatinine_phosphokinase"]].mean()
df2 = df.groupby('diabetes')[['sex',  'smoking']].count()

df.merge(df2, how = 'left', on = 'diabetes')
```

같은 행끼리 병합: join()

    - 데이터프레임 + 데이터프레임 또는 시리즈와 행을 중심으로 병합할 때 사용
    - 기본구조: DataFrame.join(other, on = None, how='left')
        - other: 병합하고자 하는 데이터프레임이나 시리즈
        - on: 병합의 기준이 되는 열이나 인덱스(지정하지 않으면 기본= 인덱스)
        - how: 병합방식을 지정(left=왼쪽 기준으로, right=오른쪽 기준으로 병합)


```python
patient_info = pd.DataFrame({
    'Patient_ID':['P123','P456','P789'],
    'Name': ['조태욱', '박태경', '배채원']
})

hospital_info = pd.DataFrame({
     'Patient_ID':['P123','P456','P789'],
     'Hospital':['내과','피부과','정형외과']
})

result = patient_info.set_index('Patient_ID').join(hospital_info.set_index('Patient_ID'))
result
```

▶ 데이터 합치기: concat()

        - s1의 모든 행 (상위 5개) 후 s2의 모든 행 (하위5개) 가 연결 됨

```python
s2 = df[['age','diabetes','sex','DEATH_EVENT']].tail(5)
s2
```

## 데이터프레임의 행 / 열에 함수를 적용
### apply() 함수로 특정 열에 함수 적용
- apply() 함수는 데이터프레임 특정 열의 각 요소에 함수를 적용
- 새로운 열에 반환된 결과를 저장

예시)
df_hf 데이터프레임에 age 열의 각 값에 1을 더한값을 새로운 age2 라는 새로운 열에 저장 가능함

▶ 단일 열에 함수 적용하기


```python
#age값에 +1하여 age2에 저장
df['age2'] = df['age'].apply(lambda x: x+1)
df[['age', 'age2']].head(3)
```

## zip() 함수
여러개의 리스트나 튜플을 받아서 각 **리스트나 튜플의 동일한 위치에 있는 **요소들끼리 **튜플로 묶어**준다.

        예시)
        id  = [1, 2, 3]
        df_hf 데이터프레임의 age, smoking, sex
        상위 3개 값들이 저장된 변수들을 zip() 함수를 사용하여
        각 리스트에 동일한 위치에 있는 값들을 순차적으로 튜플로 묶는다.
             ➞ 3개의 튜플 ( id , age , smoking , sex )

```python
id  = [1, 2, 3]

#df_hf(hear DataFrame)
age = df['age'].head(3)
smoking = df['smoking'].head(3)
sex = df['sex'].head(3)

record  = zip (id, age, smoking, sex)
print(list(record))
```

▶ dict 함수로 딕셔너리 생성

순서대로 저장되어있는 튜플들은 딕셔너리로 변환

dict 함수로 2개의 객체를 묶어 키-값 쌍을 생성
        
        예시)
        id =[ 1, 2, 3]
        name = [a, b, c, d]
        - 순서대로 묶어 zip을 통해 각 요소를 키와 값으로 튜플을 생성할수 있습니다:
            (1, 'a'), (2, 'b'), (3, 'c'), (4, 'd').
        - 생성된 튜플들을 dict 함수로 전달하여 딕셔너리를 생성할수 있습니다.

```python
id =[ 1, 2, 3]
name = ['a', 'b', 'c', 'd']
record_dict = dict(zip(id,  name))
print(record_dict)
```

## map() 함수

기본구조: map(function, iterable)
- 리스트나 시리즈에 사용
- 한 값을 다른 값에 대응시켜줄때 사용

    예시)
    0은 비흡연자, 1은 흡연자로 바꾸고 싶다면 df_hf 데이터프레임의 흡연자 'smoking' 과 비흡연자 'non-smoking' 열의 값들이 얼마나 있는지 카운트하기 위해 value_counts() 를 사용한 후,
    map 함수로 값을 변경하여 0= 'non-smoking' , 1= 'smoking' 로 바꿔줄 수 있다.




```python
#흡연자와 비흡연자 열의 값 확인
print(df['smoking'].value_counts())
```

```python
#smlking정보 중 상위 3명 정보 가져오기
smoking = df['smoking'].head(3)
smoking
```

```python
smoking.map({0:'non-smoking', 1:'smoking'})
```
