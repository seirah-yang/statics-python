pandas 데이터 구조 :
 -  '시리즈' : 1차원 배열로 단일 데이터 타입 (예: 정수, 실수, 문자열)을 저장하고 처리한다.
- '데이터프레임': 2차원 구조, 다차원의 데이터(정수, 실수, 문자연, python객체 등)을 저장하고 처리하는데 사용하며, 열과 행으로 이루엉져 있다.

pandas 역할 : 대량의 데이터를 표준화된 형식으로 처리하고 저장 가능

병원같은 의료기관에서는 환자의 나이, 성별, 질병, 검사 결과 등을 다루는데에 높은 장점이 있다.

```python
# 시리즈 데이터 구조 : 시리즈 생성
import pandas as pd

#index값 미지정 시리즈 생성 : 기본값 0
patient_weight = pd.Series ([48,89,50,72,88,84,90,101,103,60])
print(patient_weight)
```

```python
# 인덱스 값 지정된 시리즈 생성:
patient_weight = pd.Series ([48,89,50,72,88,84,90,101,103,60],
index = ['환자1','환자2','환자3','환자4','환자5','환자6','환자7', '환자8','환자9','환자10'])

print(patient_weight)
```

```python
#열:   height,  weight,  age
#행:  pt 1-10
data = {'height':[170,168,172,175,169,181,174,178,185,172],
        'weight':[70,65,80,75,68,77,83,73,88,76],
        'age':[33,45,24,44,37,29,61,28,35,47]}

patients_df = pd.DataFrame(data, index = ['환자1','환자2','환자3', '환자4', '환자5', '환자6', '환자7', '환자8', '환자9', '환자10'])

print(patient_df)
```

```python
# 새로운 데이터 계산 후 추가
patients_df['BMI'] = patients_df['weight'] /((patients_df)['height']/100**2)

print(patients_df)
```
